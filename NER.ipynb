```python
import nltk
from nltk.tokenize import word_tokenize, regexp_tokenize
from nltk.tag import pos_tag

from pprint import pprint
```

```python
s = 'Google’s India and South East Asia President Rajan Anandan has put in his papers after an eight-year stint at the American tech giant.\n\nMr. Rajan, also an investor in multiple startups, is slated to join venture fund Sequoia Capital. He will work with Google till the end of this month. Vikas Agnihotri will take on the interim responsibility. He is currently serving as Google’s Country Director for Sales.\n\n"We are grateful to Rajan for his huge contribution to Google over the past eight years. His entrepreneurial zeal and leadership has helped grow the overall internet ecosystem in India and Southeast Asia,” Google President Asia Pacific Scott Beaumont said in an emailed statement. Mr. Rajan had joined Google from Microsoft India in 2010. Previously, he was also been associated with Dell India and McKinsey & Company.\n\nIn a tweet, Mr. Rajan said: “8 amazing years. 850 million internet users across India and SEA. Many billions of revenue and fastest growing region in the world. Incredible team that thinks big and executes superbly. Thank you @GoogleIndia #GoogleSEA. Loved every minute.”\n\nSome of companies in which Mr. Rajan has invested include WebEngage, Instamojo, Capillary Technologies, TravelKhana, and EasyGov. Sequoia Capital Managing Director Shailendra J Singh, in a Linkedin post, said Mr. Rajan will join the leadership team at the firm, in addition to the six current managing directors.\n\nMr. Rajan will focus on developing ‘Surge’ programme for startups by acting as an investment advisor and mentor to the founders. Surge aims to engineer rapid early progress for startups, by enabling disproportionate access to capital, talent, network and decades of company-building knowledge.\n\n“Rajan’s deep understanding of technology, significant operating expertise and track record of growing tech businesses across the region will help Surge founders scale and build the transformational businesses of tomorrow,” Mr. Singh said.''
print(s)
```

```python
tokens = regexp_tokenize(s, pattern='\w+')
tags = pos_tag(tokens)
pprint(tags)
```

```python
# https://towardsdatascience.com/named-entity-recognition-with-nltk-and-spacy-8c4a7d88e7da
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(tags)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
from nltk import ne_chunk
ne_tree = ne_chunk(pos_tag(word_tokenize(s)))
print(ne_tree)
```

```python
import spacy
from spacy import displacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(s)
pprint([(X.text, X.label_) for X in doc.ents])
```

```python
pprint([(X, X.ent_iob_, X.ent_type_) for X in doc])
```

```python
labels = [x.label_ for x in doc.ents]
Counter(labels)
```

```python
items = [x.text for x in doc.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in doc.sents]
print(sentences[20])
```

```python
displacy.render(nlp(str(sentences[20])), jupyter=True, style='ent')
```

```python
displacy.render(nlp(str(sentences[20])), style='dep', jupyter = True, options = {'distance': 120})
```

```python
dict([(str(x), x.label_) for x in nlp(str(sentences[20])).ents])
```

```python
print([(x, x.ent_iob_, x.ent_type_) for x in sentences[20]])
```

```python
# sentiment analyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
print("Sentiment score: ", sia.polarity_scores(s)['compound'])
```
